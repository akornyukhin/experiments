```python
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
random_numbers = [random.randint(0, 1_000_000)/10_000 for _ in range(1_000_000)]

df = pd.DataFrame(random_numbers, columns=["random_numbers"])
df["rounded"] = df["random_numbers"].round(2)
df["rounded_diff"] = df["rounded"] - df["random_numbers"]
df["rounded_diff_cumsum"] = df["rounded_diff"].cumsum()

plt.figure(figsize=(10, 6))
plt.plot(df["rounded_diff_cumsum"], label="Cumulative Difference")
plt.axhline(0, color='black', linewidth=0.5)
```

```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))

#  
for i in range(100):
    random_numbers = [random.randint(0, 1_000_000) / 10_000 for _ in range(1_000_000)]
    df = pd.DataFrame(random_numbers, columns=["random_numbers"])
    df["rounded"] = df["random_numbers"].round(2)
    df["rounded_diff"] = df["rounded"] - df["random_numbers"]
    df["rounded_diff_cumsum"] = df["rounded_diff"].cumsum()
    plt.plot(df["rounded_diff_cumsum"], alpha=0.5, linewidth=0.8, label=f"Run {i+1}" if i == 0 else None)

# ---     ---
n = np.arange(1, len(df) + 1)
sigma = 0.005 / np.sqrt(3)          #    
upper = 2 * sigma * np.sqrt(n)      #   ( 95%)
lower = -upper                      #  

plt.plot(upper, 'k--', linewidth=1.5, label='+2n')
plt.plot(lower, 'k--', linewidth=1.5, label='-2n')

# ---  ---
plt.axhline(0, color='black', linewidth=0.8)
plt.title("Cumulative Rounding Differences Across 100 Series\nwith 95% Confidence Envelope")
plt.xlabel("Index")
plt.ylabel("Cumulative Difference")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
